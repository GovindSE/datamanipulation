```python
df["yearadded"]= df["Date"].dt.year
df['monthadded']= df['Date'].dt.month
titlesperyear= df["yearadded"].value_counts()
titlesperyear
```

```python
import calendar
titlepermonth= df['monthadded'].value_counts().sort_index()

titlepermonth.index= titlepermonth.index.astype(int).map(lambda x:calendar.month_name[x])
print(titlepermonth)
```

```python
movies_df = df[df['type'] == 'Movie']
movies_df['year_month'] = movies_df['Date'].dt.to_period('M')
topmovies= movies_df['year_month'].value_counts()
topmovies.sort_index()
```

```python
tv=df[(df["type"]== "TV Show") & (df["rating"]=="TV-MA")]
tv
```

```python
movies_df = df[df['type'] == 'Movie']
movies_df['duration_minutes'] = movies_df['duration'].astype(str).str.extract(r'(\d+)').astype(float)
avg_duration_by_country = movies_df.groupby('country')['duration_minutes'].mean().sort_values(ascending=False)
#print(avg_duration_by_country.head())
print(movies_df.groupby('country')['duration_minutes'])

```

```python
avg_duration_by_year = movies_df.groupby('release_year')['duration_minutes'].mean().sort_index()
print(avg_duration_by_year.tail())
```
